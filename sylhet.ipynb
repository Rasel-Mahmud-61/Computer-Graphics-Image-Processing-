```python
import matplotlib.pyplot as plt

# Line drawing algorithm
def draw_line(x1, y1, x2, y2):
    is_steep = abs(y2 - y1) > abs(x2 - x1)
    if is_steep:
        x1, y1 = y1, x1
        x2, y2 = y2, x2
    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    p = 2 * dy - dx
    x_values = []
    y_values = []
    x = x1
    y = y1
    while True:
        if is_steep:
            x_values.append(y)
            y_values.append(x)
        else:
            x_values.append(x)
            y_values.append(y)
        if x == x2 and y == y2:
            break
        if p < 0:
            p += 2 * dy
        else:
            y += sy
            p += 2 * dy - 2 * dx
        x += sx
    return x_values, y_values

# "S" shape
s_values = []
s_values.append(draw_line(-60, 10, -40, 10))  # Top horizontal line
s_values.append(draw_line(-60, 10, -60, 0))   # Left vertical line
s_values.append(draw_line(-60, 0, -40, 0))    # Middle horizontal line
s_values.append(draw_line(-40, 0, -40, -10))  # Right vertical line
s_values.append(draw_line(-60, -10, -40, -10)) # Bottom horizontal line

# "Y" shape
y_values = []
y_values.append(draw_line(-30, 10, -20, 0))   # Left diagonal line for "Y"
y_values.append(draw_line(-20, 0, -10, 10))   # Right diagonal line for "Y"
y_values.append(draw_line(-20, 0, -20, -10))  # Vertical line for "Y"

# "L" shape
l_values = []
l_values.append(draw_line(0, 10, 0, -10))     # Left vertical line
l_values.append(draw_line(0, -10, 10, -10))   # Bottom horizontal line

# "H" shape
h_values = []
h_values.append(draw_line(20, 10, 20, -10))   # Left vertical line
h_values.append(draw_line(30, 10, 30, -10))   # Right vertical line
h_values.append(draw_line(20, 0, 30, 0))      # Middle horizontal line

# "E" shape
e_values = []
e_values.append(draw_line(40, 10, 55, 10))    # Top horizontal line
e_values.append(draw_line(40, 0, 50, 0))      # Middle horizontal line
e_values.append(draw_line(40, -10, 55, -10))  # Bottom horizontal line
e_values.append(draw_line(40, 10, 40, -10))   # Left vertical line

# "T" shape
t_values = []
t_values.append(draw_line(65, 10, 85, 10))    # Top horizontal line
t_values.append(draw_line(75, 10, 75, -10))   # Vertical line for "T"

# Combining all line values
all_values = s_values + y_values + l_values + h_values + e_values + t_values

# Unpacking
x_values, y_values = zip(*all_values)
x_values = [item for sublist in x_values for item in (sublist + [None])]
y_values = [item for sublist in y_values for item in (sublist + [None])]

# Plotting function
def plot_func():
    plt.plot(x_values, y_values, color='black')
    plt.title("SYLHET")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.axis("equal")
    plt.show()

plot_func()
```
