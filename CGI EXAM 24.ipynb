```python
import matplotlib.pyplot as plt

# Line drawing algorithm
def draw_line(x1, y1, x2, y2):
    is_steep = abs(y2 - y1) > abs(x2 - x1)
    if is_steep:
        x1, y1 = y1, x1
        x2, y2 = y2, x2
    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    p = 2 * dy - dx
    x_values = []
    y_values = []
    x, y = x1, y1
    while True:
        if is_steep:
            x_values.append(y)
            y_values.append(x)
        else:
            x_values.append(x)
            y_values.append(y)
        if x == x2 and y == y2:
            break
        if p < 0:
            p += 2 * dy
        else:
            y += sy
            p += 2 * dy - 2 * dx
        x += sx
    return x_values, y_values

# Define the shapes for each character in "CGIEXAM24"
# Adjusting the positions and sizes of each letter

# "C" shape
c_values = []
c_values.append(draw_line(-40, 10, -20, 10))   # Top
c_values.append(draw_line(-40, 10, -40, -10))  # Left vertical
c_values.append(draw_line(-40, -10, -20, -10)) # Bottom

# "G" shape
g_values = []
g_values.append(draw_line(-10, 10, 10, 10))    # Top
g_values.append(draw_line(-10, 10, -10, -10))  # Left vertical
g_values.append(draw_line(-10, -10, 10, -10))  # Bottom
g_values.append(draw_line(10, -10, 10, 0))     # Right vertical
g_values.append(draw_line(0, 0, 10, 0))        # Middle line

# "I" shape
i_values = []
i_values.append(draw_line(20, 10, 20, -10))    # Vertical line for "I"

# "E" shape
e_values = []
e_values.append(draw_line(30, 10, 50, 10))     # Top horizontal
e_values.append(draw_line(30, 0, 45, 0))       # Middle horizontal
e_values.append(draw_line(30, -10, 50, -10))   # Bottom horizontal
e_values.append(draw_line(30, 10, 30, -10))    # Left vertical

# "X" shape
x_values = []
x_values.append(draw_line(60, 10, 80, -10))    # Diagonal down
x_values.append(draw_line(60, -10, 80, 10))    # Diagonal up

# "A" shape
a_values = []
a_values.append(draw_line(90, -10, 100, 10))   # Left diagonal
a_values.append(draw_line(100, 10, 110, -10))  # Right diagonal
a_values.append(draw_line(95, 0, 105, 0))      # Middle horizontal

# "M" shape
m_values = []
m_values.append(draw_line(120, -10, 120, 10))  # Left vertical
m_values.append(draw_line(120, 10, 130, 0))    # Left diagonal
m_values.append(draw_line(130, 0, 140, 10))    # Right diagonal
m_values.append(draw_line(140, 10, 140, -10))  # Right vertical

# "2" shape
two_values = []
two_values.append(draw_line(150, 10, 170, 10)) # Top horizontal
two_values.append(draw_line(170, 10, 170, 0))  # Right vertical
two_values.append(draw_line(150, 0, 170, 0))   # Middle horizontal
two_values.append(draw_line(150, 0, 150, -10)) # Bottom left vertical
two_values.append(draw_line(150, -10, 170, -10)) # Bottom horizontal

# "4" shape
four_values = []
four_values.append(draw_line(180, 10, 180, 0)) # Left vertical
four_values.append(draw_line(180, 0, 190, 0))  # Middle horizontal
four_values.append(draw_line(190, 10, 190, -10)) # Right vertical

# Combine all values for CGIEXAM24
all_values = c_values + g_values + i_values + e_values + x_values + a_values + m_values + two_values + four_values

# Unpack values
x_values, y_values = zip(*all_values)
x_values = [item for sublist in x_values for item in (sublist + [None])]
y_values = [item for sublist in y_values for item in (sublist + [None])]

# Plotting function
def plot_func():
    plt.plot(x_values, y_values, color='black')
    plt.title("CGIEXAM24")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.axis("equal")
    plt.show()

# Call function to plot
plot_func()
```
